```python
numbers = [-6, -4, -1, 0, 4, 5, 7, 8, 10]
for i in range(0, len(numbers)):
    print("Number at index {} is {}".format(i, numbers[i]))
```

```python
for i in range(0, len(numbers)):
    if numbers[i] % 2 == 0:
        print("Even number at index {} is {}".format(i, numbers[i]))
    else:
        print("Odd number at index {} is {}".format(i, numbers[i]))
```

### Task 2:
Make a list that have all the integers from 0 to 99 that are divisible by 4 but
not divisible by 6.

Hint: `my_list.append(a)` adds `a` to the end of the `my_list`.

```python
result = []
for number in range(0, 100):
    if number % 4 == 0 and number % 6 != 0:
        result.append(number)

print(result)
```
