```python
import matplotlib.pyplot as plt
%matplotlib inline

t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

fig, ax = plt.subplots()
ax.plot(t, s)
ax.set(xlabel='Hours studied (s)', ylabel='voltage (mV)', title='Voltage vs Time')
ax.grid()
plt.show()
```
